Primero vamos a empezar definiendo la señal x(t), y la simplificamos de una mera diferente usando una identidad trigonometrica


```python
import numpy as np # importamos las librerias de uso numpy para arrelgos
import matplotlib.pyplot as plt # para graficar
from scipy.spatial.distance import cdist
from scipy.integrate import quad  # Para integrar numéricamente
# usamos codigos separados para correjir los errores mas facilmente
```

```python
A = 1 # amplitud de la señal
F0 = 1 # frecuencia Fundamental
T = 1/ F0 # Periodo Fundamnetal
W0 = 2*np.pi*F0 # Frecuencia Angular fundamental
Numero_armonicos = 10 # numero de armonicos a usar
t = np.linspace (-T,T,1000) # vector de tiempo
# defino la señal usando la identidad trigonometrica
x_t = A**2 * (1+np.cos(4*np.pi*F0*t))/2
```

NOTA aqui va foto del calculo usando la identidad trigonometrica

```python
# grafico mi señal
plt.figure(figsize=(10, 6))
plt.plot(t, x_t, label="$x(t) = |A \cos(2 \pi F_0 t)|^2$", linewidth=4)
plt.title("Signal $x(t)$")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
