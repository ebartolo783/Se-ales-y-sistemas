```python
%%writefile 3_📊_Punto_3.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Configuración de la página
st.set_page_config(page_title="Análisis de THD", page_icon="📊")
st.title("Análisis de THD en Rectificadores")
st.markdown("### Definiciones de Conceptos Base:")
st.markdown("Abstracto: Esta medida de distorsión tiene dos definiciones diferentes, la primera nos dice que se debe tomar como punto de referencia el armónico fundamental (THDF) y la segunda se debe tomar como comparación a la raíz cuadrada de la media de la señal (THDR). Para evitar ambigüedades, errores o malas interpretaciones se optará por utilizar la primera definición, puesto que es la que la IEEE recomienda seguir, además que es la definición adoptada por la IEC.")

st.markdown("Distorsion: Cuando el voltaje o la corriente de un sistema eléctrico tienen deformaciones con respecto a la forma de onda sinusoidal, se dice que la señal está distorsionada. Dicha distorsión puede deberse a fenómenos transitorios como arranque de motores, conmutación de capacitores, efectos de tormentas o fallas por cortocircuito entre otras.")

st.markdown("""Para que se considere como distorsión armónica las deformaciones en una señal, se deben de cumplir las siguientes condiciones:

Que la señal tenga valores definidos dentro del intervalo, lo que implica que la energía contenida es finita

Que la señal sea periódica, teniendo la misma forma de onda en cada ciclo de la señal de corriente o voltaje.

Que la señal sea periódica, teniendo la misma forma de onda en cada ciclo de la señal de corriente o voltaje.""")
st.markdown("Distorsion total de armonicos (TDH):La distorsión total de armónicos (THD) Es el parámetro de medición de distorsión más conocido, por lo que es recomendable para medir la distorsión en parámetros individuales (I y V), dicha medición consiste básicamente en comparar el armónico fundamental, con todos los demás armónicos de la señal")
st.markdown("Otra forma para calcular el THD es haciendo uso de la transformada rápida de Fourier (FFT), ya que esta herramienta nos permite analizar las frecuencias de una señal descomponiéndola en sus componentes sinusoidales y cosenoidales, lo que nos permite extraer los valores de la amplitud en cada componente y de esta manera compararlo con la fundamental")

st.markdown("THD y el factor de potencia (FP): El factor de potencia a grosso modo es un índice que mide la eficiencia energética de un sistema eléctrico y la forma más común de calcularlo que se suele enseñar es mediante el coseno del angulo formado entre la potencia activa y la potencia aparente, Esta definición, aunque es correcta, su validez se limita a señales que no posean armónicos distorsionados y es que esta fórmula solo tiene en cuenta la información de la fundamental, lo que puede generar problemas en los equipos a la larga ya que los armónicos tienen la capacidad de producir una serie de problemas, como la resonancia en los bancos de capacitores, quemar fusibles, accionar los interruptores diferenciales y termo magnéticos, aumentar gradualmente la corriente que pasa por el neutro, etc.. Para solucionar este problema se desarrolló el Factor de potencia distorsionante, el cual hace uso del THD, para de esta manera tomar en cuenta todo el espectro de la señal")

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Parámetros del sistema
sampling_rate = 10000  # Frecuencia de muestreo
T = 1.0 / sampling_rate  # Intervalo de tiempo
t = np.linspace(0, 1.0, sampling_rate)

# Función para generar la señal del rectificador de onda completa
def rectificador_onda_completa(t, f, R, C=None):
    signal = np.abs(np.sin(2 * np.pi * f * t))
    if C is not None:
        tau = R * C
        signal = signal * np.exp(-t / tau)
    return signal

# Función para calcular la FFT
def calcular_fft(signal, sampling_rate):
    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1 / sampling_rate)[:N // 2]
    return xf, np.abs(yf[:N // 2])

# Función para calcular THD
def calcular_thd(yf):
    V1 = yf[1]  # Primer armónico
    harmonics = np.sqrt(np.sum(yf[2:]**2))
    thd = harmonics / V1
    return thd

# Interfaz de usuario en Streamlit
st.title("Análisis de THD en rectificador de onda completa")

# Sliders para R y C
R = st.slider("Valor de la resistencia (ohmios)", 1.0, 1000.0, 100.0)
C = st.slider("Valor del condensador (faradios)", 1e-6, 1e-3, 1e-5, format="%.6f")

# Frecuencia de la señal de entrada
f = 60  # Hz

# Generar señales
signal_resistiva = rectificador_onda_completa(t, f, R)
signal_rc = rectificador_onda_completa(t, f, R, C)

# Calcular FFT
xf_res, yf_res = calcular_fft(signal_resistiva, sampling_rate)
xf_rc, yf_rc = calcular_fft(signal_rc, sampling_rate)

# Calcular THD
thd_resistiva = calcular_thd(yf_res)
thd_rc = calcular_thd(yf_rc)

st.write(f"THD para carga resistiva pura: {thd_resistiva:.4f}")
st.write(f"THD para carga RC: {thd_rc:.4f}")

# Graficar señales y su FFT
fig, axs = plt.subplots(2, 2, figsize=(12, 6))

axs[0, 0].plot(t[:500], signal_resistiva[:500])
axs[0, 0].set_title("Señal con carga resistiva pura")
axs[0, 0].set_xlabel("Tiempo [s]")
axs[0, 0].set_ylabel("Amplitud")

axs[0, 1].plot(t[:500], signal_rc[:500])
axs[0, 1].set_title("Señal con carga RC")
axs[0, 1].set_xlabel("Tiempo [s]")
axs[0, 1].set_ylabel("Amplitud")

axs[1, 0].plot(xf_res, yf_res)
axs[1, 0].set_title("FFT - Carga resistiva pura")
axs[1, 0].set_xlabel("Frecuencia [Hz]")
axs[1, 0].set_ylabel("Amplitud")

axs[1, 1].plot(xf_rc, yf_rc)
axs[1, 1].set_title("FFT - Carga RC")
axs[1, 1].set_xlabel("Frecuencia [Hz]")
axs[1, 1].set_ylabel("Amplitud")

plt.tight_layout()
st.pyplot(fig)




```

```python
%%writefile 4_📊_Punto_4.py
import os
import zipfile
import streamlit as st
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE

# ------------------------- 1. Función para guardar ZIP -------------------------
def save_uploaded_file(uploaded_file):
    zip_path = os.path.join("results", "audios.zip")
    with open(zip_path, "wb") as f:
        f.write(uploaded_file.getbuffer())
    return zip_path

# ------------------------- 2. Función para extraer ZIP -------------------------
def extract_zip(zip_path, extract_to="results/audios"):
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_to)

# ------------------------- 3. Clasificación de audios -------------------------
def classify_audios(base_path="results/audios/"):
    data = []
    if not os.path.exists(base_path):
        return pd.DataFrame(columns=["NOMBRE DEL ARCHIVO", "ESTUDIANTE", "CLAVE"])

    for folder in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder)
        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                if file.endswith(".wav"):
                    clave = 1 if folder == "correcto" else 2
                    data.append([file, folder, clave])

    return pd.DataFrame(data, columns=["NOMBRE DEL ARCHIVO", "ESTUDIANTE", "CLAVE"])

# ------------------------- 4. INTERFAZ: Subida y Clasificación -------------------------
st.title("Subida y Clasificación de Audios")

uploaded_file = st.file_uploader("Sube un archivo ZIP con audios", type=["zip"])

if uploaded_file:
    zip_path = save_uploaded_file(uploaded_file)
    extract_zip(zip_path)
    st.success("Archivos extraídos correctamente en 'results/audios/'")

    df = classify_audios()
    st.write("### Audios Clasificados")
    st.dataframe(df)

# ------------------------- 5. Recorte y Normalización de Frecuencia -------------------------
st.title("Normalización de Frecuencia y Recorte de Audios")

output_folder = "results/audios_recortados/"
os.makedirs(output_folder, exist_ok=True)

duraciones = []
fs_list = []
audio_data = []

for folder in ["correcto", "incorrecto"]:
    folder_path = os.path.join("results/audios", folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            if file.endswith(".wav"):
                audio_path = os.path.join(folder_path, file)
                y, sr = librosa.load(audio_path, sr=None)
                fs_list.append(sr)
                duraciones.append(librosa.get_duration(y=y, sr=sr))
                audio_data.append((y, sr, audio_path))

# Encontrar la menor frecuencia de muestreo para estandarizar
target_fs = min(fs_list) if fs_list else None

duracion_media = np.mean(duraciones) if duraciones else 0

# Normalización de frecuencia y recorte
audios_recortados = []
max_samples = int(duracion_media * target_fs) if target_fs else 0

for y, sr, audio_path in audio_data:
    if sr != target_fs:
        y = librosa.resample(y, orig_sr=sr, target_sr=target_fs)
    if len(y) > max_samples:
        y = y[:max_samples]
    else:
        y = np.pad(y, (0, max_samples - len(y)), mode="constant")
    output_path = os.path.join(output_folder, os.path.basename(audio_path))
    sf.write(output_path, y, target_fs)
    audios_recortados.append(output_path)

# ------------------------- 6. Análisis de Fourier -------------------------
st.title("Análisis de Fourier")

if audios_recortados:
    x_t = []
    for audio_path in audios_recortados:
        y, _ = librosa.load(audio_path, sr=target_fs)
        x_t.append(y)

    x_t = np.array(x_t)
    ts = max_samples / target_fs
    vf = np.fft.rfftfreq(x_t.shape[1], 1 / target_fs)
    Xw = np.fft.rfft(x_t, axis=1)
    Xw_mag = np.abs(Xw)

    # Frecuencia máxima
    f_max = vf[np.argmax(Xw_mag, axis=1)]
    st.write(f"Frecuencia máxima detectada: {max(f_max):.2f} Hz")

    # Forma de onda de todos los audios
    fig, ax = plt.subplots(figsize=(10, 4))
    for i in range(x_t.shape[0]):
        ax.plot(np.linspace(0, ts, x_t.shape[1]), x_t[i], alpha=0.5)
    ax.set_xlabel("Tiempo (s)")
    ax.set_ylabel("Amplitud")
    ax.set_title("Forma de onda de todos los audios")
    st.pyplot(fig)

    # Espectros normalizados
    sca = MinMaxScaler()
    Xw_ = sca.fit_transform(Xw_mag.T).T

    fig, ax = plt.subplots(figsize=(10, 4))
    for i in range(Xw_.shape[0]):
        ax.plot(vf, Xw_[i])
    ax.set_xlabel("f[Hz]")
    ax.set_ylabel("|X(f)|")
    ax.set_title("Espectros Normalizados")
    st.pyplot(fig)

    # Espectros en decibeles
    fig, ax = plt.subplots(figsize=(10, 4))
    for i in range(Xw_.shape[0]):
        ax.plot(vf, 20 * np.log10(Xw_[i] + 1e-10))
    ax.set_xlabel("f[Hz]")
    ax.set_ylabel("|X(f)| dB")
    ax.set_title("Espectros en dB")
    st.pyplot(fig)

else:
    st.error("No se encontraron archivos de audio recortados.")

import os
import numpy as np
import librosa
import matplotlib.pyplot as plt
import streamlit as st
from scipy.spatial.distance import cdist
from io import BytesIO

# Configuración de la aplicación Streamlit
st.title("Clasificación de Audio con Transformada de Fourier")

# Cargar el nuevo archivo de audio
uploaded_file = st.file_uploader("Sube un archivo de audio", type=["wav"])

if uploaded_file is not None:
    y_new, sr_new = librosa.load(uploaded_file, sr=48000)
    Xw_new = np.abs(np.fft.rfft(y_new))

    # Cargar los datos previos
    base_path = "results/audios_recortados/"
    estudiantes = {"correcto": 1, "incorrecto": 2}

    wav_files = []
    labels = []

    for folder, label in estudiantes.items():
        folder_path = os.path.join(base_path, folder)
        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                if file.endswith(".wav"):
                    wav_files.append(os.path.join(folder_path, file))
                    labels.append(label)

    # Procesar Fourier para los audios previos
    Xw_list = []
    for file_path in wav_files:
        y, sr = librosa.load(file_path, sr=48000)
        Xw = np.abs(np.fft.rfft(y))
        Xw_list.append(Xw)

    # Determinar el tamaño máximo y ajustar todas las transformadas a esta longitud
    max_length = max(map(len, Xw_list))
    Xw_array = np.array([
        np.pad(Xw, (0, max_length - len(Xw)), mode='constant') if len(Xw) < max_length else Xw[:max_length]
        for Xw in Xw_list
    ])

    # Ajustar el nuevo audio al mismo tamaño
    Xw_new_padded = np.pad(Xw_new, (0, max_length - len(Xw_new)), mode='constant') if len(Xw_new) < max_length else Xw_new[:max_length]

    # Calcular la distancia entre el nuevo audio y los audios previos
    distances = cdist(Xw_array, Xw_new_padded.reshape(1, -1), metric="euclidean")

    # Encontrar el índice del audio más cercano
    closest_index = np.argmin(distances)
    predicted_label = labels[closest_index]

    # Mostrar el resultado
    if predicted_label == 1:
        st.success("Audio aceptado (correcto - Azul)")
    else:
        st.error("Audio no aceptado (incorrecto - Amarillo)")

    # Graficar las distancias
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.bar(range(len(distances)), distances.flatten(), color="blue", alpha=0.6)
    ax.axvline(closest_index, color="red", linestyle="--", label="Más cercano")
    ax.set_xlabel("Muestras de referencia")
    ax.set_ylabel("Distancia euclidiana")
    ax.legend()

    st.pyplot(fig)







```
