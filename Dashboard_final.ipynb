## **1. Punto_1**


```python
%%writefile 1_üìà_Punto_1.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

# Configurar la p√°gina de Streamlit
st.set_page_config(page_title="Dashboard - Se√±ales y Sistemas", page_icon="üìä")
st.title("Punto 1: Se√±al Espectro y Reconstruccion")
st.markdown("""
Primero se mostrara la se√±al en el tiempo y luego su espectro.
""")
# Definir par√°metros
To = np.pi  # Periodo
Fo = 1 / To  # Frecuencia fundamental
Fs = 300 * Fo  # Frecuencia de muestreo
tv = np.arange(-To / 2, To / 2, 1 / Fs)  # Vector de tiempo
Nm = len(tv)  # N√∫mero de muestras
A = 1  # Amplitud de la se√±al
x = (A * np.cos(2 * np.pi * Fo * tv)) ** 2  # Se√±al

# Graficar se√±al
st.subheader("Se√±al en el tiempo")
fig, ax = plt.subplots()
ax.plot(tv, x, 'b', linewidth=2)
ax.grid()
ax.set_xlabel("t [s]")
ax.set_ylabel("x(t)")
st.pyplot(fig)

# Calcular coeficientes de Fourier
N = 50  # N√∫mero de arm√≥nicos
wo = 2 * np.pi / To  # Frecuencia angular fundamental
nv = np.linspace(-N, N, 2 * N + 1)  # Vector de arm√≥nicos
Cn = np.zeros(2 * N + 1, dtype=np.complex_)
Cn[N] = (A ** 2) / 2  # Nivel DC
Cn[N + 2] = (A ** 2) / 4
Cn[N - 2] = (A ** 2) / 4

# Graficar espectro de Fourier
st.subheader("Espectro de Fourier")
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].stem(nv, np.real(Cn), 'b')
axs[0, 0].set_xlabel("n")
axs[0, 0].set_ylabel("Re{C_n}")
axs[0, 0].grid()

axs[0, 1].stem(nv, np.imag(Cn), 'b')
axs[0, 1].set_xlabel("n")
axs[0, 1].set_ylabel("Im{C_n}")
axs[0, 1].grid()

axs[1, 0].stem(nv, abs(Cn), 'b')
axs[1, 0].set_xlabel("n")
axs[1, 0].set_ylabel("|C_n|")
axs[1, 0].grid()

axs[1, 1].stem(nv, np.angle(Cn), 'b')
axs[1, 1].set_xlabel("n")
axs[1, 1].set_ylabel("Fase")
axs[1, 1].grid()

fig.tight_layout()
st.pyplot(fig)
st.markdown("los siguientes graficos muestran el diagrama de Bode (magnitud y fase ) en decibeles")
# Graficar espectro en dB y fase en grados
fig, axs = plt.subplots(2, 1, figsize=(8, 6))
axs[0].stem(nv, 20 * np.log10(abs(Cn)), 'b')
axs[0].set_xlabel("n")
axs[0].set_ylabel("|C_n| [dB]")
axs[0].grid()

axs[1].stem(nv, np.angle(Cn, deg=True), 'b')
axs[1].set_xlabel("n")
axs[1].set_ylabel("Fase [¬∞]")
axs[1].grid()

fig.tight_layout()
st.pyplot(fig)

# Control deslizante para reconstrucci√≥n de la se√±al
st.subheader("Reconstrucci√≥n de la se√±al con arm√≥nicos")
Na = st.slider("N√∫mero de arm√≥nicos", min_value=0, max_value=N, value=1)

# Estimaci√≥n de la se√±al reconstruida
Px = 0.375
ind = np.arange(N - Na, N + Na + 1)
er = 1 - np.sum(abs(Cn[ind]) ** 2) / Px
phin = np.zeros((Nm, 2 * N + 1), dtype=np.complex_)
for n in range(-N, N + 1):
    phin[:, n + N] = np.exp(1j * n * wo * tv)
xe = phin[:, ind] @ Cn[ind]

fig, ax = plt.subplots()
ax.plot(tv, xe, 'b', marker='o', markersize=3, label='$x_e(t)$')
ax.plot(tv, x, 'r', label='x(t)')
ax.set_title(f'Error relativo: {100 * er:.2f}%')
ax.set_xlabel("t [s]")
ax.set_ylabel("x(t)")
ax.grid()
ax.legend()
st.pyplot(fig)

st.markdown("En este link se podran observar los calculos realizados a mano del calculo de las constantes[solucion teorica](https://github.com/ebartolo783/Se-ales-y-sistemas/tree/main/CalculosParcial2")



# Conclusi√≥n

st.markdown("""
### Resumen
- Se gener√≥ la se√±al \( x(t) = \cos^2(2\pi f_0 t) \).
- Se mostr√≥ la se√±al en el dominio del tiempo y su espectro en el dominio de la frecuencia.
- Se implement√≥ una reconstrucci√≥n con n√∫mero variable de arm√≥nicos.
""")
```

```python
%%writefile 2_üåç_Punto_2.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.io.wavfile import write
import io
from scipy.io.wavfile import write
import tempfile

# Configuraci√≥n de p√°gina
st.set_page_config(page_title="Transformada de Fourier - AM", page_icon="üìä")
st.title("Transformada de Fourier de una se√±al modulada AM")



st.markdown("### Expresi√≥n general")
st.latex(r"Y(\omega) = \mathcal{F}\{y(t)\} = \mathcal{F}\left\{\left(1+\frac{m(t)}{A_c}\right)c(t)\right\}")
st.latex(r"Y(\omega) = \mathcal{F}\{c(t)\} + \frac{1}{A_c}\mathcal{F}\{m(t)c(t)\}")

st.markdown("### Transformada de la portadora")
st.latex(r"C(\omega) = \mathcal{F}\{c(t)\} = \mathcal{F}\{A_c\sin(2\pi F_c t)\}")
st.latex(r"\mathcal{F}\{e^{\pm j\omega_0 t}\} = 2\pi\delta(\omega \mp \omega_0)")
st.latex(r"C(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right)")

st.markdown("### Transformada del t√©rmino modulado")
st.latex(r"\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \mathcal{F}\{m(t)\sin(2\pi F_c t)\}")
st.latex(r"\mathcal{F}\{x(t)e^{\pm j\omega_0t}\} = X(\omega \mp \omega_0)")
st.latex(r"\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)")

st.markdown("### Espectro final de la se√±al modulada")
st.latex(r"Y(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right) + \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)")


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

st.title("Transformada de Fourier de la Se√±al Modulada")

# Par√°metros de la se√±al
Fc = 100  # Frecuencia de la portadora
Fm = 10   # Frecuencia del mensaje
Am = 1    # Amplitud del mensaje
ti = 0
tf = 5 / Fm  # Cantidad de per√≠odos a graficar
Fs = 5 * Fc  # Frecuencia de muestreo

# C√°lculo de se√±ales
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo
m = Am * np.cos(2 * np.pi * Fm * t)  # Mensaje
Im = 0.8  # √çndice de modulaci√≥n
Ac = max(abs(m)) / Im  # Amplitud de la portadora
c = Ac * np.sin(2 * np.pi * Fc * t)  # Portadora

st.write(f"Amplitud de la portadora calculada: {Ac}")

# Graficar se√±ales
fig, ax = plt.subplots()
ax.plot(t, c, label='Portadora')
ax.plot(t, m, label='Mensaje')
ax.legend()
ax.set_xlabel("Tiempo (s)")
ax.set_ylabel("Amplitud")
ax.set_title("Se√±ales de la portadora y el mensaje")

st.pyplot(fig)

# Graficar se√±al modulada
y = (1 + m / Ac) * c  # AM
fig_mod, ax_mod = plt.subplots()
ax_mod.plot(t, y, label='Se√±al Modulada (AM)')
ax_mod.legend()
ax_mod.set_xlabel("Tiempo (s)")
ax_mod.set_ylabel("Amplitud")
ax_mod.set_title("Se√±al Modulada en AM")
st.pyplot(fig_mod)

st.markdown("Aca se visualizara el grafico del espectro a partir de las se√±ales mensaje portadora y modulada, con su respectivo color indicando cual es cada una")

# C√°lculo de la FFT
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)
vfre = np.fft.fftfreq(len(c), 1 / Fs)

# Graficar espectro
fig_fft, ax_fft = plt.subplots()
ax_fft.plot(vfre, abs(Xfy), label='Modulado')
ax_fft.plot(vfre, abs(Xfm), label='Mensaje')
ax_fft.plot(vfre, abs(Xfc), label='Portadora')
ax_fft.legend()
ax_fft.set_xlabel("Frecuencia [Hz]")
ax_fft.set_ylabel("|X(f)|")
ax_fft.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax_fft.set_title("Espectro de Frecuencia de las Se√±ales")
st.pyplot(fig_fft)

import yt_dlp
import soundfile as sf
import streamlit as st
import subprocess

st.markdown("### Descarga de Audio y Procesamiento")
st.markdown("Aca se uso la cancion de salsa tiutalda Un verano en Nueva York del Gran combo donde se uso el fragmento de la cancion del segundo 20 hasta el segundo 25")
# Descargar audio de YouTube
def download_audio(link):
    ydl_opts = {
        'format': 'bestaudio/best',  # Descargar solo el mejor audio disponible
        'outtmpl': 'audio.mp3',  # Guardar el archivo de audio como audio.mp3
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([link])

link = "https://www.youtube.com/watch?v=hd0cl2d2ziU"
download_audio(link)

# Convertir el audio descargado a WAV si es necesario
subprocess.run(["ffmpeg", "-y", "-i", "audio.mp3", "output.wav"])

# Leer el archivo de audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)
st.write(f"Frecuencia de muestreo: {fs} Hz")
st.write(f"Forma del audio: {x.shape}")

# Tomar solo un canal si es est√©reo
if len(x.shape) > 1:
    x = x[:, 0]

# Reproducir segmento de audio
start_time = 20
end_time = 25
audio_segment = x[int(fs * start_time):int(fs * end_time)]

sf.write("temp_segment.wav", audio_segment, fs)
st.audio("temp_segment.wav")



# Configuraci√≥n inicial de la app
st.title("Visualizaci√≥n de Se√±al de Audio")

# Par√°metros
fs = 1000  # Frecuencia de muestreo en Hz
ti = 20  # Tiempo inicial en segundos
tf = 25  # Tiempo final en segundos

# Crear se√±al de ejemplo
np.random.seed(0)
x = np.random.randn((tf - ti) * fs, 1)  # Simulaci√≥n de una se√±al de audio

# Copiar datos para procesamiento
xpro = x.copy()

# Extraer segmento de la se√±al
xs = xpro[:len(xpro), 0]  # Asegurar longitud correcta

# Crear vector de tiempo con la misma longitud que xs
tt = np.linspace(ti, tf, len(xs), endpoint=False)

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(tt, xs, label="Canal izquierdo", linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("x(t)")
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)
len(xs)



# Configuraci√≥n inicial de la app
st.title("Modulaci√≥n AM - Se√±al Portadora y Mensaje")

# Entrada del usuario para el √≠ndice de modulaci√≥n
Im = st.number_input("Ingrese el √≠ndice de modulaci√≥n:", min_value=0.1, max_value=2.0, value=1.0, step=0.1)

# Frecuencia portadora y muestreo
Fc = 1000  # Frecuencia de portadora
ti = 20
tf = 25  # Cantidad de periodos a graficar
Fs = 48000  # Frecuencia de muestreo

# Crear vector de tiempo
t = np.arange(ti, tf, 1 / Fs)

# Generar se√±al de mensaje de ejemplo
np.random.seed(0)
xs = np.random.randn(len(t))  # Simulaci√≥n de una se√±al de audio

# Calcular amplitud de la portadora
Ac = max(abs(xs)) / Im
c = Ac * np.sin(2 * np.pi * Fc * t)  # Se√±al portadora

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, c, label='Portadora', linewidth=1)
ax.plot(t, xs, label='Mensaje', linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("Amplitud")
ax.set_ylim([-10,10])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)

# Mostrar valor de Ac
st.write(f"Amplitud de la portadora (Ac): {Ac:.2f}")
# Se√±al modulada en AM
y = (1 + xs / Ac) * c

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, y, label='Se√±al Modulada AM', linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("y(t)")
ax.set_ylim([-2,2])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)

# Configuraci√≥n inicial de la app
st.title("Espectro de Se√±ales - Portadora")

# Frecuencia de muestreo
Fs = 48000  # Frecuencia de muestreo

# Calcular espectros usando la Transformada R√°pida de Fourier (RFFT)
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)

# Generar vector de frecuencias
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Graficar espectros
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(vfre, abs(Xfc), label='Portadora', linewidth=1)
ax.set_xlabel("Frecuencia [Hz]")
ax.set_ylabel("|X(f)|")
ax.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)


# Configuraci√≥n inicial de la app
st.title("Espectro de Se√±ales - Modulada y Mensaje")

# Frecuencia de muestreo
Fs = 48000  # Frecuencia de muestreo

# Calcular espectros usando la Transformada R√°pida de Fourier (RFFT)
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)

# Generar vector de frecuencias
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Crear figura con dos subgr√°ficos
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# Primer subplot - Espectro de la se√±al modulada
axes[0].plot(vfre, abs(Xfy), label='Modulado', color='r')
axes[0].set_xlabel("Frecuencia [Hz]")
axes[0].set_ylabel("|X(f)|")
axes[0].set_ylim([0, 5000])
axes[0].legend()
axes[0].grid(True)

# Segundo subplot - Espectro de la se√±al mensaje
axes[1].plot(vfre, abs(Xfm), label='Mensaje')
axes[1].set_xlabel("Frecuencia [Hz]")
axes[1].set_ylabel("|X(f)|")
axes[1].set_ylim([0, 5000])
axes[1].legend()
axes[1].grid(True)

# Ajustar dise√±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)

# Funci√≥n para guardar y reproducir audio en Streamlit
def play_audio(signal, sample_rate, label):
    with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as tmpfile:
        write(tmpfile.name, sample_rate, signal.astype(np.float32))
        st.audio(tmpfile.name, format="audio/wav", start_time=0)
        st.write(label)

# Mostrar los audios
tabs = st.tabs(["Portadora", "Mensaje", "Modulada"])
with tabs[0]:
    play_audio(c, Fs, "Audio de la portadora")
with tabs[1]:
    play_audio(xs, Fs, "Audio del mensaje")
with tabs[2]:
    play_audio(y, Fs, "Audio de la se√±al modulada")

```

```python
%%writefile 3_üìä_Punto_3.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Configuraci√≥n de la p√°gina
st.set_page_config(page_title="An√°lisis de THD", page_icon="üìä")
st.title("An√°lisis de THD en Rectificadores")
st.markdown("### Definiciones de Conceptos Base:")
st.markdown("Abstracto: Esta medida de distorsi√≥n tiene dos definiciones diferentes, la primera nos dice que se debe tomar como punto de referencia el arm√≥nico fundamental (THDF) y la segunda se debe tomar como comparaci√≥n a la ra√≠z cuadrada de la media de la se√±al (THDR). Para evitar ambig√ºedades, errores o malas interpretaciones se optar√° por utilizar la primera definici√≥n, puesto que es la que la IEEE recomienda seguir, adem√°s que es la definici√≥n adoptada por la IEC.")

st.markdown("Distorsion: Cuando el voltaje o la corriente de un sistema el√©ctrico tienen deformaciones con respecto a la forma de onda sinusoidal, se dice que la se√±al est√° distorsionada. Dicha distorsi√≥n puede deberse a fen√≥menos transitorios como arranque de motores, conmutaci√≥n de capacitores, efectos de tormentas o fallas por cortocircuito entre otras.")

st.markdown("""Para que se considere como distorsi√≥n arm√≥nica las deformaciones en una se√±al, se deben de cumplir las siguientes condiciones:

Que la se√±al tenga valores definidos dentro del intervalo, lo que implica que la energ√≠a contenida es finita

Que la se√±al sea peri√≥dica, teniendo la misma forma de onda en cada ciclo de la se√±al de corriente o voltaje.

Que la se√±al sea peri√≥dica, teniendo la misma forma de onda en cada ciclo de la se√±al de corriente o voltaje.""")
st.markdown("Distorsion total de armonicos (TDH):La distorsi√≥n total de arm√≥nicos (THD) Es el par√°metro de medici√≥n de distorsi√≥n m√°s conocido, por lo que es recomendable para medir la distorsi√≥n en par√°metros individuales (I y V), dicha medici√≥n consiste b√°sicamente en comparar el arm√≥nico fundamental, con todos los dem√°s arm√≥nicos de la se√±al")
st.markdown("Otra forma para calcular el THD es haciendo uso de la transformada r√°pida de Fourier (FFT), ya que esta herramienta nos permite analizar las frecuencias de una se√±al descomponi√©ndola en sus componentes sinusoidales y cosenoidales, lo que nos permite extraer los valores de la amplitud en cada componente y de esta manera compararlo con la fundamental")

st.markdown("THD y el factor de potencia (FP): El factor de potencia a grosso modo es un √≠ndice que mide la eficiencia energ√©tica de un sistema el√©ctrico y la forma m√°s com√∫n de calcularlo que se suele ense√±ar es mediante el coseno del angulo formado entre la potencia activa y la potencia aparente, Esta definici√≥n, aunque es correcta, su validez se limita a se√±ales que no posean arm√≥nicos distorsionados y es que esta f√≥rmula solo tiene en cuenta la informaci√≥n de la fundamental, lo que puede generar problemas en los equipos a la larga ya que los arm√≥nicos tienen la capacidad de producir una serie de problemas, como la resonancia en los bancos de capacitores, quemar fusibles, accionar los interruptores diferenciales y termo magn√©ticos, aumentar gradualmente la corriente que pasa por el neutro, etc.. Para solucionar este problema se desarroll√≥ el Factor de potencia distorsionante, el cual hace uso del THD, para de esta manera tomar en cuenta todo el espectro de la se√±al")

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Par√°metros del sistema
sampling_rate = 10000  # Frecuencia de muestreo
T = 1.0 / sampling_rate  # Intervalo de tiempo
t = np.linspace(0, 1.0, sampling_rate)

# Funci√≥n para generar se√±al del rectificador de onda completa
def rectificador_onda_completa(t, f, R, C=None):
    signal = np.abs(np.sin(2 * np.pi * f * t))
    if C is not None:
        tau = R * C
        signal = signal * np.exp(-t / tau)
    return signal

# Calcular la FFT y los arm√≥nicos
def calcular_fft(signal, sampling_rate):
    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1 / sampling_rate)[:N // 2]
    return xf, np.abs(yf[:N // 2])

# Calcular THD
def calcular_thd(yf):
    V1 = yf[1]  # Primer arm√≥nico
    harmonics = np.sqrt(np.sum(yf[2:]**2))
    return harmonics / V1

# Interfaz en Streamlit
st.title("An√°lisis de Rectificador de Onda Completa")

# Entradas del usuario
f = st.number_input("Frecuencia de la se√±al de entrada (Hz)", value=60, min_value=1)
R = st.number_input("Valor de la resistencia (Ohmios)", value=1000, min_value=1)
C = st.number_input("Valor del condensador (Faradios)", value=0.001, min_value=0.000001, format="%.6f")

# Generar se√±ales
signal_resistiva = rectificador_onda_completa(t, f, R)
signal_rc = rectificador_onda_completa(t, f, R, C)

# Calcular FFT
tf_res, yf_res = calcular_fft(signal_resistiva, sampling_rate)
tf_rc, yf_rc = calcular_fft(signal_rc, sampling_rate)

# Calcular THD
thd_resistiva = calcular_thd(yf_res)
thd_rc = calcular_thd(yf_rc)

st.write(f"**THD para carga resistiva pura:** {thd_resistiva:.4f}")
st.write(f"**THD para carga RC:** {thd_rc:.4f}")

# Graficar se√±ales y su espectro
fig, axs = plt.subplots(2, 2, figsize=(12, 6))

axs[0, 0].plot(t[:500], signal_resistiva[:500])
axs[0, 0].set_title("Se√±al con carga resistiva pura")
axs[0, 0].set_xlabel("Tiempo [s]")
axs[0, 0].set_ylabel("Amplitud")

axs[0, 1].plot(t[:500], signal_rc[:500])
axs[0, 1].set_title("Se√±al con carga RC")
axs[0, 1].set_xlabel("Tiempo [s]")
axs[0, 1].set_ylabel("Amplitud")

axs[1, 0].plot(tf_res, yf_res)
axs[1, 0].set_title("FFT - Carga resistiva pura")
axs[1, 0].set_xlabel("Frecuencia [Hz]")
axs[1, 0].set_ylabel("Amplitud")

axs[1, 1].plot(tf_rc, yf_rc)
axs[1, 1].set_title("FFT - Carga RC")
axs[1, 1].set_xlabel("Frecuencia [Hz]")
axs[1, 1].set_ylabel("Amplitud")

plt.tight_layout()
st.pyplot(fig)



```
