```python
%%writefile 2_ðŸŒ_Punto_2.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy.io.wavfile import write
import io
from scipy.io.wavfile import write
import tempfile

# ConfiguraciÃ³n de pÃ¡gina
st.set_page_config(page_title="Transformada de Fourier - AM", page_icon="ðŸ“Š")
st.title("Transformada de Fourier de una seÃ±al modulada AM")



st.markdown("### ExpresiÃ³n general")
st.latex(r"Y(\omega) = \mathcal{F}\{y(t)\} = \mathcal{F}\left\{\left(1+\frac{m(t)}{A_c}\right)c(t)\right\}")
st.latex(r"Y(\omega) = \mathcal{F}\{c(t)\} + \frac{1}{A_c}\mathcal{F}\{m(t)c(t)\}")

st.markdown("### Transformada de la portadora")
st.latex(r"C(\omega) = \mathcal{F}\{c(t)\} = \mathcal{F}\{A_c\sin(2\pi F_c t)\}")
st.latex(r"\mathcal{F}\{e^{\pm j\omega_0 t}\} = 2\pi\delta(\omega \mp \omega_0)")
st.latex(r"C(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right)")

st.markdown("### Transformada del tÃ©rmino modulado")
st.latex(r"\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \mathcal{F}\{m(t)\sin(2\pi F_c t)\}")
st.latex(r"\mathcal{F}\{x(t)e^{\pm j\omega_0t}\} = X(\omega \mp \omega_0)")
st.latex(r"\frac{1}{A_c}\mathcal{F}\{m(t)c(t)\} = \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)")

st.markdown("### Espectro final de la seÃ±al modulada")
st.latex(r"Y(\omega) = \frac{A_c\pi}{j}\left(\delta(\omega - 2\pi F_c) - \delta(\omega + 2\pi F_c)\right) + \frac{1}{2j}\left(M(\omega - 2\pi F_c) - M(\omega + 2\pi F_c)\right)")


import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

st.title("Transformada de Fourier de la SeÃ±al Modulada")

# ParÃ¡metros de la seÃ±al
Fc = 100  # Frecuencia de la portadora
Fm = 10   # Frecuencia del mensaje
Am = 1    # Amplitud del mensaje
ti = 0
tf = 5 / Fm  # Cantidad de perÃ­odos a graficar
Fs = 5 * Fc  # Frecuencia de muestreo

# CÃ¡lculo de seÃ±ales
t = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo
m = Am * np.cos(2 * np.pi * Fm * t)  # Mensaje
Im = 0.8  # Ãndice de modulaciÃ³n
Ac = max(abs(m)) / Im  # Amplitud de la portadora
c = Ac * np.sin(2 * np.pi * Fc * t)  # Portadora

st.write(f"Amplitud de la portadora calculada: {Ac}")

# Graficar seÃ±ales
fig, ax = plt.subplots()
ax.plot(t, c, label='Portadora')
ax.plot(t, m, label='Mensaje')
ax.legend()
ax.set_xlabel("Tiempo (s)")
ax.set_ylabel("Amplitud")
ax.set_title("SeÃ±ales de la portadora y el mensaje")

st.pyplot(fig)

# Graficar seÃ±al modulada
y = (1 + m / Ac) * c  # AM
fig_mod, ax_mod = plt.subplots()
ax_mod.plot(t, y, label='SeÃ±al Modulada (AM)')
ax_mod.legend()
ax_mod.set_xlabel("Tiempo (s)")
ax_mod.set_ylabel("Amplitud")
ax_mod.set_title("SeÃ±al Modulada en AM")
st.pyplot(fig_mod)

st.markdown("Aca se visualizara el grafico del espectro a partir de las seÃ±ales mensaje portadora y modulada, con su respectivo color indicando cual es cada una")

# CÃ¡lculo de la FFT
Xfc = np.fft.fft(c)
Xfm = np.fft.fft(m)
Xfy = np.fft.fft(y)
vfre = np.fft.fftfreq(len(c), 1 / Fs)

# Graficar espectro
fig_fft, ax_fft = plt.subplots()
ax_fft.plot(vfre, abs(Xfy), label='Modulado')
ax_fft.plot(vfre, abs(Xfm), label='Mensaje')
ax_fft.plot(vfre, abs(Xfc), label='Portadora')
ax_fft.legend()
ax_fft.set_xlabel("Frecuencia [Hz]")
ax_fft.set_ylabel("|X(f)|")
ax_fft.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax_fft.set_title("Espectro de Frecuencia de las SeÃ±ales")
st.pyplot(fig_fft)

import yt_dlp
import soundfile as sf
import streamlit as st
import subprocess

st.markdown("### Descarga de Audio y Procesamiento")
st.markdown("Aca se uso la cancion de salsa tiutalda Un verano en Nueva York del Gran combo donde se uso el fragmento de la cancion del segundo 20 hasta el segundo 25")
# Descargar audio de YouTube
def download_audio(link):
    ydl_opts = {
        'format': 'bestaudio/best',  # Descargar solo el mejor audio disponible
        'outtmpl': 'audio.mp3',  # Guardar el archivo de audio como audio.mp3
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([link])

link = "https://www.youtube.com/watch?v=hd0cl2d2ziU"
download_audio(link)

# Convertir el audio descargado a WAV si es necesario
subprocess.run(["ffmpeg", "-y", "-i", "audio.mp3", "output.wav"])

# Leer el archivo de audio
nombre_out = "output.wav"
x, fs = sf.read(nombre_out)
st.write(f"Frecuencia de muestreo: {fs} Hz")
st.write(f"Forma del audio: {x.shape}")

# Tomar solo un canal si es estÃ©reo
if len(x.shape) > 1:
    x = x[:, 0]

# Reproducir segmento de audio
start_time = 20
end_time = 25
audio_segment = x[int(fs * start_time):int(fs * end_time)]

sf.write("temp_segment.wav", audio_segment, fs)
st.audio("temp_segment.wav")



# ConfiguraciÃ³n inicial de la app
st.title("VisualizaciÃ³n de SeÃ±al de Audio")

# ParÃ¡metros
fs = 1000  # Frecuencia de muestreo en Hz
ti = 20  # Tiempo inicial en segundos
tf = 25  # Tiempo final en segundos

# Crear seÃ±al de ejemplo
np.random.seed(0)
x = np.random.randn((tf - ti) * fs, 1)  # SimulaciÃ³n de una seÃ±al de audio

# Copiar datos para procesamiento
xpro = x.copy()

# Extraer segmento de la seÃ±al
xs = xpro[:len(xpro), 0]  # Asegurar longitud correcta

# Crear vector de tiempo con la misma longitud que xs
tt = np.linspace(ti, tf, len(xs), endpoint=False)

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(tt, xs, label="Canal izquierdo", linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("x(t)")
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)
len(xs)



# ConfiguraciÃ³n inicial de la app
st.title("ModulaciÃ³n AM - SeÃ±al Portadora y Mensaje")

# Entrada del usuario para el Ã­ndice de modulaciÃ³n
Im = st.number_input("Ingrese el Ã­ndice de modulaciÃ³n:", min_value=0.1, max_value=2.0, value=1.0, step=0.1)

# Frecuencia portadora y muestreo
Fc = 1000  # Frecuencia de portadora
ti = 20
tf = 25  # Cantidad de periodos a graficar
Fs = 48000  # Frecuencia de muestreo

# Crear vector de tiempo
t = np.arange(ti, tf, 1 / Fs)

# Generar seÃ±al de mensaje de ejemplo
np.random.seed(0)
xs = np.random.randn(len(t))  # SimulaciÃ³n de una seÃ±al de audio

# Calcular amplitud de la portadora
Ac = max(abs(xs)) / Im
c = Ac * np.sin(2 * np.pi * Fc * t)  # SeÃ±al portadora

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, c, label='Portadora', linewidth=1)
ax.plot(t, xs, label='Mensaje', linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("Amplitud")
ax.set_ylim([-10,10])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)

# Mostrar valor de Ac
st.write(f"Amplitud de la portadora (Ac): {Ac:.2f}")
# SeÃ±al modulada en AM
y = (1 + xs / Ac) * c

# Graficar
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, y, label='SeÃ±al Modulada AM', linewidth=1)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("y(t)")
ax.set_ylim([-2,2])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)

# ConfiguraciÃ³n inicial de la app
st.title("Espectro de SeÃ±ales - Portadora")

# Frecuencia de muestreo
Fs = 48000  # Frecuencia de muestreo

# Calcular espectros usando la Transformada RÃ¡pida de Fourier (RFFT)
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)

# Generar vector de frecuencias
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Graficar espectros
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(vfre, abs(Xfc), label='Portadora', linewidth=1)
ax.set_xlabel("Frecuencia [Hz]")
ax.set_ylabel("|X(f)|")
ax.set_ylim([min(abs(Xfm)), 1.1 * max(abs(Xfm))])
ax.legend()
ax.grid(True)

# Mostrar en Streamlit
st.pyplot(fig)


# ConfiguraciÃ³n inicial de la app
st.title("Espectro de SeÃ±ales - Modulada y Mensaje")

# Frecuencia de muestreo
Fs = 48000  # Frecuencia de muestreo

# Calcular espectros usando la Transformada RÃ¡pida de Fourier (RFFT)
Xfc = np.fft.rfft(c)
Xfm = np.fft.rfft(xs)
Xfy = np.fft.rfft(y)

# Generar vector de frecuencias
vfre = np.fft.rfftfreq(len(c), 1 / Fs)

# Crear figura con dos subgrÃ¡ficos
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# Primer subplot - Espectro de la seÃ±al modulada
axes[0].plot(vfre, abs(Xfy), label='Modulado', color='r')
axes[0].set_xlabel("Frecuencia [Hz]")
axes[0].set_ylabel("|X(f)|")
axes[0].set_ylim([0, 5000])
axes[0].legend()
axes[0].grid(True)

# Segundo subplot - Espectro de la seÃ±al mensaje
axes[1].plot(vfre, abs(Xfm), label='Mensaje')
axes[1].set_xlabel("Frecuencia [Hz]")
axes[1].set_ylabel("|X(f)|")
axes[1].set_ylim([0, 5000])
axes[1].legend()
axes[1].grid(True)

# Ajustar diseÃ±o y mostrar en Streamlit
fig.tight_layout()
st.pyplot(fig)

# FunciÃ³n para guardar y reproducir audio en Streamlit
def play_audio(signal, sample_rate, label):
    with tempfile.NamedTemporaryFile(delete=False, suffix=".wav") as tmpfile:
        write(tmpfile.name, sample_rate, signal.astype(np.float32))
        st.audio(tmpfile.name, format="audio/wav", start_time=0)
        st.write(label)

# Mostrar los audios
tabs = st.tabs(["Portadora", "Mensaje", "Modulada"])
with tabs[0]:
    play_audio(c, Fs, "Audio de la portadora")
with tabs[1]:
    play_audio(xs, Fs, "Audio del mensaje")
with tabs[2]:
    play_audio(y, Fs, "Audio de la seÃ±al modulada")

```

```python
%%writefile 3_ðŸ“Š_Punto_3.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ConfiguraciÃ³n de la pÃ¡gina
st.set_page_config(page_title="AnÃ¡lisis de THD", page_icon="ðŸ“Š")
st.title("AnÃ¡lisis de THD en Rectificadores")
st.markdown("### Definiciones de Conceptos Base:")
st.markdown("Abstracto: Esta medida de distorsiÃ³n tiene dos definiciones diferentes, la primera nos dice que se debe tomar como punto de referencia el armÃ³nico fundamental (THDF) y la segunda se debe tomar como comparaciÃ³n a la raÃ­z cuadrada de la media de la seÃ±al (THDR). Para evitar ambigÃ¼edades, errores o malas interpretaciones se optarÃ¡ por utilizar la primera definiciÃ³n, puesto que es la que la IEEE recomienda seguir, ademÃ¡s que es la definiciÃ³n adoptada por la IEC.")

st.markdown("Distorsion: Cuando el voltaje o la corriente de un sistema elÃ©ctrico tienen deformaciones con respecto a la forma de onda sinusoidal, se dice que la seÃ±al estÃ¡ distorsionada. Dicha distorsiÃ³n puede deberse a fenÃ³menos transitorios como arranque de motores, conmutaciÃ³n de capacitores, efectos de tormentas o fallas por cortocircuito entre otras.")

st.markdown("""Para que se considere como distorsiÃ³n armÃ³nica las deformaciones en una seÃ±al, se deben de cumplir las siguientes condiciones:

Que la seÃ±al tenga valores definidos dentro del intervalo, lo que implica que la energÃ­a contenida es finita

Que la seÃ±al sea periÃ³dica, teniendo la misma forma de onda en cada ciclo de la seÃ±al de corriente o voltaje.

Que la seÃ±al sea periÃ³dica, teniendo la misma forma de onda en cada ciclo de la seÃ±al de corriente o voltaje.""")
st.markdown("Distorsion total de armonicos (TDH):La distorsiÃ³n total de armÃ³nicos (THD) Es el parÃ¡metro de mediciÃ³n de distorsiÃ³n mÃ¡s conocido, por lo que es recomendable para medir la distorsiÃ³n en parÃ¡metros individuales (I y V), dicha mediciÃ³n consiste bÃ¡sicamente en comparar el armÃ³nico fundamental, con todos los demÃ¡s armÃ³nicos de la seÃ±al")
st.markdown("Otra forma para calcular el THD es haciendo uso de la transformada rÃ¡pida de Fourier (FFT), ya que esta herramienta nos permite analizar las frecuencias de una seÃ±al descomponiÃ©ndola en sus componentes sinusoidales y cosenoidales, lo que nos permite extraer los valores de la amplitud en cada componente y de esta manera compararlo con la fundamental")

st.markdown("THD y el factor de potencia (FP): El factor de potencia a grosso modo es un Ã­ndice que mide la eficiencia energÃ©tica de un sistema elÃ©ctrico y la forma mÃ¡s comÃºn de calcularlo que se suele enseÃ±ar es mediante el coseno del angulo formado entre la potencia activa y la potencia aparente, Esta definiciÃ³n, aunque es correcta, su validez se limita a seÃ±ales que no posean armÃ³nicos distorsionados y es que esta fÃ³rmula solo tiene en cuenta la informaciÃ³n de la fundamental, lo que puede generar problemas en los equipos a la larga ya que los armÃ³nicos tienen la capacidad de producir una serie de problemas, como la resonancia en los bancos de capacitores, quemar fusibles, accionar los interruptores diferenciales y termo magnÃ©ticos, aumentar gradualmente la corriente que pasa por el neutro, etc.. Para solucionar este problema se desarrollÃ³ el Factor de potencia distorsionante, el cual hace uso del THD, para de esta manera tomar en cuenta todo el espectro de la seÃ±al")

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ParÃ¡metros del sistema
sampling_rate = 10000  # Frecuencia de muestreo
T = 1.0 / sampling_rate  # Intervalo de tiempo
t = np.linspace(0, 1.0, sampling_rate)

# FunciÃ³n para generar seÃ±al del rectificador de onda completa
def rectificador_onda_completa(t, f, R, C=None):
    signal = np.abs(np.sin(2 * np.pi * f * t))
    if C is not None:
        tau = R * C
        signal = signal * np.exp(-t / tau)
    return signal

# Calcular la FFT y los armÃ³nicos
def calcular_fft(signal, sampling_rate):
    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1 / sampling_rate)[:N // 2]
    return xf, np.abs(yf[:N // 2])

# Calcular THD
def calcular_thd(yf):
    V1 = yf[1]  # Primer armÃ³nico
    harmonics = np.sqrt(np.sum(yf[2:]**2))
    return harmonics / V1

# Interfaz en Streamlit
st.title("AnÃ¡lisis de Rectificador de Onda Completa")

# Entradas del usuario
f = st.number_input("Frecuencia de la seÃ±al de entrada (Hz)", value=60, min_value=1)
R = st.number_input("Valor de la resistencia (Ohmios)", value=1000, min_value=1)
C = st.number_input("Valor del condensador (Faradios)", value=0.001, min_value=0.000001, format="%.6f")

# Generar seÃ±ales
signal_resistiva = rectificador_onda_completa(t, f, R)
signal_rc = rectificador_onda_completa(t, f, R, C)

# Calcular FFT
tf_res, yf_res = calcular_fft(signal_resistiva, sampling_rate)
tf_rc, yf_rc = calcular_fft(signal_rc, sampling_rate)

# Calcular THD
thd_resistiva = calcular_thd(yf_res)
thd_rc = calcular_thd(yf_rc)

st.write(f"**THD para carga resistiva pura:** {thd_resistiva:.4f}")
st.write(f"**THD para carga RC:** {thd_rc:.4f}")

# Graficar seÃ±ales y su espectro
fig, axs = plt.subplots(2, 2, figsize=(12, 6))

axs[0, 0].plot(t[:500], signal_resistiva[:500])
axs[0, 0].set_title("SeÃ±al con carga resistiva pura")
axs[0, 0].set_xlabel("Tiempo [s]")
axs[0, 0].set_ylabel("Amplitud")

axs[0, 1].plot(t[:500], signal_rc[:500])
axs[0, 1].set_title("SeÃ±al con carga RC")
axs[0, 1].set_xlabel("Tiempo [s]")
axs[0, 1].set_ylabel("Amplitud")

axs[1, 0].plot(tf_res, yf_res)
axs[1, 0].set_title("FFT - Carga resistiva pura")
axs[1, 0].set_xlabel("Frecuencia [Hz]")
axs[1, 0].set_ylabel("Amplitud")

axs[1, 1].plot(tf_rc, yf_rc)
axs[1, 1].set_title("FFT - Carga RC")
axs[1, 1].set_xlabel("Frecuencia [Hz]")
axs[1, 1].set_ylabel("Amplitud")

plt.tight_layout()
st.pyplot(fig)



```
